```python
#Group by city and calculate the average cost for each city
city_rating_asc = df.groupby('City')['Average Cost for two'].mean().sort_values(ascending=True)
city_rating_asc
```

```python
city_rating_desc = df.groupby('City')['Average Cost for two'].mean().sort_values(ascending=False)
city_rating_desc
```

```python
city_rating_desc_top5 = df.groupby('City')['Average Cost for two'].mean().sort_values(ascending=False).head()
city_rating_desc_top5
```

```python
#Plotting it on the graph
city_rating = df.groupby('City')['Average Cost for two'].mean().sort_values(ascending=False).head().plot(kind='bar', figsize=(12,6))
plt.title('Average Cost for Two by City')
plt.xlabel('City')
plt.ylabel('Average Cost for Two')
plt.xticks(rotation=45)
plt.show()
```

```python
#Cusines with the highest average cost for two
cuisine_rating_desc = df.groupby('Cuisines')['Average Cost for two'].mean().sort_values(ascending=False).head()
cuisine_rating_desc.plot(kind='pie', figsize=(20,6))
plt.title('Average Cost for Two by Cuisine')    
plt.xlabel('Cuisine')
plt.xticks(rotation=45)
plt.legend(title='Cuisine', loc='upper right')
plt.show()
```

```python
#Cusines total number of votes 
df.groupby('Cuisines')['Votes'].sum().sort_values(ascending=False).head()
```

```python
#PLotting the graph
cuisine_votes_desc = df.groupby('Cuisines')['Votes'].sum().sort_values(ascending=False).head()
cuisine_votes_desc.plot(kind='line', figsize=(12,6), color='green')  
plt.title('Total Votes by Cuisine')
plt.xlabel('Cuisine')
plt.ylabel('Total Votes')
plt.xticks(rotation=45)
plt.show()
```
